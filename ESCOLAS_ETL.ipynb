```python
import pandas as pd
import numpy as np
```

```python
escolas_2010 = pd.read_csv('data/ESCOLAS.csv', delimiter='|', low_memory=False)
```

```python
escolas_2010.head()
```

```python
escolas_ce_2010 = escolas_2010[(escolas_2010['SIGLA'] == 'CE') & (escolas_2010['DESC_SITUACAO_FUNCIONAMENTO'] == 1)]
```

```python
escolas_ce_2010.reset_index(inplace=True)
escolas_ce_2010.head()
```

```python
for c in escolas_ce_2010.columns:
    print(c)
```

```python
escolas_ce_2010_s = escolas_ce_2010[['PK_COD_ENTIDADE', 'NO_ENTIDADE', 'NUM_FUNCIONARIOS', 'ID_DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'ID_BIBLIOTECA', 'ID_LABORATORIO_INFORMATICA', 'NUM_SALAS_UTILIZADAS', 'FK_COD_MUNICIPIO']]
```

```python
escolas_ce_2010_s.head()
```

```python
escolas_ce_2010_s.columns = ['COD_ESCOLA', 'NOME_ESCOLA', 'NUMERO_FUNCIONARIOS', 'DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'BIBLIOTECA', 'LABORATORIO_INFORMATICA', 'NUMERO_SALAS', 'COD_MUNICIPIO']
```

```python
escolas_ce_2010_s.to_csv('base/ESCOLAS_2010.csv', index=False)
```

```python
escolas_by_municipio = escolas_ce_2010_s.groupby(['COD_MUNICIPIO'], as_index=False)['COD_ESCOLA'].count()
```

```python
escolas_by_municipio.columns = ['COD_MUNICIPIO', 'TOT_ESCOLAS']
```

```python
escolas_by_municipio.to_csv('base/ESCOLAS_BY_MUNICIPIO.csv', index=False)
```
